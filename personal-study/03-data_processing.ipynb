```python
import pandas as pd
import seaborn as sns
```

### Column

```python
# 예제 데이터프레임 생성
data = {
    'CustomerID': [101, 102, 103, 104],
    'OrderDate': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04'],
    'Product': ['Apple', 'Banana', 'Cherry', 'Date'],
    'Quantity': [5, 3, 4, 2],
    'Price': [100,200,250,50]
}

df = pd.DataFrame(data)
display(df)
```

```python
# 컬럼 이름 변경
df.rename(columns={'CustomerID' : 'ID'}, inplace=True)
display(df)
```

```python
# OrderDate 컬럼 삭제
df.drop('OrderDate', axis=1, inplace=True)
display(df)
```

```python
# TotalAmount 추가
df['TotalAmount'] = df['Quantity'] * df['Price']
display(df)
```

### Filtering

```python
dft = sns.load_dataset('titanic')
display(dft.head())
```

```python
# 1등급석(First)의 30세 이상 승객 데이터 필터링
dft_1 = dft[(dft['class']=='First') & (dft['age']>=30)]

dft_1['age'].min() # 결과 30.0
dft_1['class'].value_counts() # 결과 First 131, Second 0 , Thirt 0
```

### merge

```python
employees = pd.DataFrame({
    'employee_id': ['E01', 'E02', 'E03', 'E04'],
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'department': ['HR', 'Engineering', 'IT', 'Marketing']
})

salaries = pd.DataFrame({
    'id': ['E01', 'E02', 'E03'],
    'salary': [70000, 85000, 95000]
})

display(employees)
display(salaries)
```

```python
# id기준 merge
pd.merge(employees, salaries
         , left_on = 'employee_id', right_on = 'id'
         , how = 'inner')[['employee_id', 'name', 'department', 'salary']]
```
