```python
import pandas as pd
```

각 월마다 시가총액(market_cap) 기준으로 1등부터 5등까지 기업 모으기
* 사용 데이터: `kospi.csv`
> Hint <br/> * For-loop<br/>* pd.merge

```python
data = pd.read_csv('kospi.csv')
```

```python
data.head()
```

```python
# trd_date 자료형 변환
data['trd_date'] = pd.to_datetime(data['trd_date'])
data['year-month'] = data['trd_date'].dt.strftime('%Y-%m')
data.head()
```

```python
list = data['year-month'].unique()

result = pd.DataFrame()

for item in list:
    tmp = data[data['year-month']==item].sort_values('market_cap', ascending=False)[['trd_date', 'co_nm', 'market_cap']].head(5)
    result = pd.concat([result, tmp], ignore_index=True)

result
```
